```python
import numpy as np
import cv2
import csv
from matplotlib import pyplot as plt
from itertools import islice
```

```python
LABELS = {
    '0': 'angry',
    '1': 'disguest',
    '2': 'fear',
    '3': 'happy',
    '4': 'sad',
    '5': 'suprise',
    '6': 'neutral'
}

SELECTED_LABELS = ('happy', 'sad', 'neutral')
```

```python
# read in the training data, row 1 - 100

SHAPE = (48, 48)
CSV_TRAIN = '../../data/train.csv'

with open(CSV_TRAIN) as csvfile:
    reader = csv.reader(csvfile, delimiter=',', quotechar='"')
    for row in islice(reader,0,100):
        t, x = row
        label = LABELS.get(t)
        if label not in SELECTED_LABELS:
            continue
        pixels = np.asarray(list(map(int, x.split()))).reshape(SHAPE)
        print(label)
        plt.imshow(pixels, cmap='gray', vmin=0, vmax=255)
        plt.show()
```
